```python
import plotly.express as px
import pandas as pd
from IPython.core.display import HTML
HTML("""
<style>
g.pointtext {display: none;}
</style>
""")
```

```python
database = pd.read_excel("spelling_project_long_12_17_2020_cleaning.xlsx")
```

```python
def plotting_3d(df):
    fig = px.scatter_3d(df, x="Levenshtein_dist", y="Cosine_dist",z = "Phono_dist",
                        color = "Spelling_Target",
                        hover_data={'Spelling_Target':True, 
                                 "Levenshtein_dist":True,
                                 "Cosine_dist":True,
                                   "Phono_dist":True,
                                   "words":True},
                       color_discrete_sequence = px.colors.qualitative.Alphabet)
    fig.update_layout(
        autosize=False,
        width=1000,
        height=600,
        margin=dict(l=20, r=20, t=30, b=80),
    )
    fig.update_traces(marker=dict(size=6))
    from IPython.core.display import HTML
    HTML("""
    <style>
    g.pointtext {display: none;}
    </style>
    """)
    fig.show()
    return fig
```

```python
total = plotting_3d(database)
```

```python
import plotly.io as pio
pio.write_html(total, file="index.html", auto_open=True)
```

```python
plotting_3d(database[database["Spelling_Target"] == "soliloquy"])
```

```python
plotting_3d(database[database["Spelling_Target"] == "separate"])
```
